# Queries

## Find all vertices in the graph

```python
%%gremlin 

g.V()
```

## This query deletes all nodes

```python
%%gremlin

g.V().drop()
```

## This finds all the results of a given process

```python
%%gremlin

g.V("lehigh6852253371").repeat(out()).until(hasLabel("result")).valueMap(true)
```

## This query finds a specific measureId from a process

```python
%%gremlin

g.V("test8035322371").repeat(out("has")).until(has("measureId","test-5-1-1-M1" )).toList()
```

## Finds the output of a specific hierarchy

```python
%%gremlin
g.V('lehigh6852253371').out("outputs").valueMap(true)
```

```python
%%gremlin

g.V('RawMat-25880397113').outE().inV().path().by('name').by(label)
```

```python
%%gremlin 
// note you have to include L afterwards because it is a long
g.V(1L)
```

# This is the query of for  traversing a specific hierarchy using the process name

```python
%%gremlin -p v,oute,inv,oute,inv,oute,inv

g.V("lehigh6852253371").outE().inV().outE().inV().outE().inV().path().by('name').by(label)
```

## Finding the Inputs of a process

```python
%%gremlin

g.V('help').properties('inputs')
```

## This query finds all the results for a given measureId in a given Process

```python
%%gremlin

g.V('blue0060132371').repeat(out()).until(has("measureId", "blue-1-1-M2"))
.out().valueMap(true)
```

## Find all the materials(inputs) that went into a process, and find the processes results

```python
%%gremlin -p v,ine,oute

g.V('lehigh6852253371').in().repeat(out()).until(hasLabel("result")).path().by('name').by(label)
```
